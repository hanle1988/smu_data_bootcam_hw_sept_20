````python
data = []

for element in elements[1:]:
    date = element.find("div", {"class": "list_date"}).text
    title = element.find("div", {"class": "content_title"}).text
    note = element.find("div", {"class": "article_teaser_body"}).text
    
    row = {}
    row["date"] = date
    row["title"] = title
    row["note"] = note
    
    data.append(row)
    
df = pd.DataFrame(data)
df
````
