```python
# Clustering/Dimensionality Reduction
from sklearn.manifold import TSNE
from sklearn.cluster import DBSCAN
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# DataScience
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read the csv
df = pd.read_csv("myopia.csv")
df
```

```python
#Drop the MYOPIC column
df.drop(columns=["MYOPIC"], inplace=True)
df.head()
```

```python
#Data needs to be numeric and not-null
df.info()
```

```python
df.AGE.value_counts()
```

```python
#Checks for obvious clustering
#Looks like top left and right will cluster

X = TSNE(learning_rate="auto").fit_transform(df)

plt.figure(figsize=(16,10))

plt.scatter(X[:, 0], X[:, 1])
plt.title("TSNE Plot")
plt.show()
```

```python
#Initialize PCA Model

pca = PCA()
pca.fit(df)

plt.plot(range(len(df.columns)), pca.explained_variance_ratio_)
plt.xlabel(" Numbers of Principal Components")
plt.ylabel("Explained Variance")
plt.title("PCA Elbow")
plt.figure(figsize=(20,12))
plt.show()
```

```python
#Elbows seem appear at 1, 2, 3 and possibly 5 (3 principal components are majority)
```

```python
list(zip(range(len(df.columns)), pca.explained_variance_ratio_))
```

```python
#There is a correlation (0 at 0.72 or 72%)
```

```python
#Summary of correlation
df.corr()
```

```python
#Principal compoments analysis
pca = PCA(n_components=3)
pca.fit(df)

```

```python
df1 = pd.DataFrame(pca.transform(df), columns=["PCA1", "PCA2", "PCA3"])
df1.head()
```

```python
# Check for obvious clusters #2.
X = TSNE(learning_rate="auto").fit_transform(df)

plt.figure(figsize=(12,8))

plt.scatter(X[:, 0], X[:, 1])
plt.title("TSNE Plot")
plt.show()
```

```python
#Does show many differences
```

```python
#Kmeans model

inertia = []
k_range = list(range(2, 20))

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df)
    
    inertia.append(kmeans.inertia_)
    
#Make the plot
plt.figure(figsize=(16,10))

plt.plot(k_range, inertia)
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Plot")

plt.show()
```

```python
#Elbows seem to appears at a bit after 2.5 and a bit before 5 (slight elbow)
```

```python
#KMeans cluster of 10

km = KMeans(n_clusters=10, random_state=42)
kmeans.fit(df)
preds = kmeans.predict(df)

plt.figure(figsize=(12,8))

plt.scatter(X[:, 0], X[:, 1], c=preds)
plt.title("TSNE Plot")
plt.show()
```

```python
##Showing overlappings, 10 seems to be over fitting
```

```python
#KMeans cluster of 8

km = KMeans(n_clusters=8, random_state=42)
kmeans.fit(df)
preds = kmeans.predict(df)

plt.figure(figsize=(12,8))

plt.scatter(X[:, 0], X[:, 1], c=preds)
plt.title("TSNE Plot")
plt.show()
```

```python
##Showing overlappings, 8 seems to be over fitting
```

```python
#KMeans cluster of 5

km = KMeans(n_clusters=5, random_state=42)
kmeans.fit(df)
preds = kmeans.predict(df)

plt.figure(figsize=(12,8))

plt.scatter(X[:, 0], X[:, 1], c=preds)
plt.title("TSNE Plot")
plt.show()
```

```python
#Purple scatters, have interesting boundaries
```

```python
#Dendrogram plot

mergings = linkage(df, method = "ward")

plt.figure(figsize=(20, 8))

dendrogram(mergings,leaf_rotation=90,leaf_font_size=5)

plt.show()
```

```python
#Green is a clear cluster
#Orange spread out into about four groups, two of which could be combined to one.
```

```python
cluster = AgglomerativeClustering(
    n_clusters=5,
    affinity="euclidean",
    linkage="ward"
)

preds = cluster.fit_predict(df)
plt.figure(figsize=(12,8))

plt.scatter(X[:, 0], X[:, 1], c=preds)
plt.title("TSNE Plot")
plt.show()
```
